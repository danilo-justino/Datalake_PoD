```python
# Importando bibliotecas

from pyspark.sql import *
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
# Leitura dos arquivos Parquet na pasta Trusted tb_fatura

tb_fatura = spark.read.parquet("Files/Refined/stage_fatura")

tb_fatura.createOrReplaceTempView("tb_fatura")

display(tb_fatura)
```

```python
%%sql
SELECT 
   COUNT(*)
FROM tb_fatura
```

```python
%%sql

-- Percentual da Distribuição do Status de Pagamento dos Clientes

SELECT
    fbc_classificacao_pagamento,
    COUNT(fbc_classificacao_pagamento) as total,
    ROUND((COUNT(fbc_classificacao_pagamento) * 100) / SUM(COUNT(fbc_classificacao_pagamento)) OVER(), 2) AS percentual
FROM tb_fatura
GROUP BY fbc_classificacao_pagamento 
ORDER BY total DESC;
```

```python
%%sql

-- Conhecendo Os valores dos Clientes SEM_PAGAMENTO

SELECT
    fbc_classificacao_pagamento,
    MIN(fvl_valor_fatura) AS vl_minimo,
    MAX(fvl_valor_fatura) AS vl_maximo,
    COUNT(fvl_valor_fatura) AS total
FROM tb_fatura
WHERE fbc_classificacao_pagamento = "SEM_PAGAMENTO"
GROUP BY fbc_classificacao_pagamento;
```

```python
%%sql

-- Criando Faixa de valor de Fatura que estão SEM_PAGAMENTO

WITH faixas AS(
    SELECT
        fbc_classificacao_pagamento,
        fvl_valor_fatura,
        CASE
            WHEN fvl_valor_fatura BETWEEN 0 AND 15000 THEN 'Fatura até 15k'
            WHEN fvl_valor_fatura <=  30000 THEN 'Fatura entre 15k a 30k'
            WHEN fvl_valor_fatura <= 45000 THEN 'Fatura entre 30k a 45k'
            WHEN fvl_valor_fatura <= 60000 THEN 'Fatura entre 45k a 60k'
            ELSE 'Fatura maior que 60k'
        END AS faixa_valor
    FROM tb_fatura
    WHERE fbc_classificacao_pagamento = 'SEM_PAGAMENTO'
)

SELECT 
    faixa_valor,
    COUNT(*) AS total_fatura,
    ROUND((COUNT(*) * 100.0) / (SELECT COUNT(*) FROM faixas), 2) AS percentual
FROM faixas
GROUP BY faixa_valor
ORDER BY total_fatura DESC;
```
